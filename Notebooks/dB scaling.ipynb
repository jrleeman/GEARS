```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.fft import rfft, rfftfreq
%matplotlib inline
```

```python
signal_frequency = 5
length_seconds = 100
noise_amplitude = 0.001
sample_freq = 100000
time = np.linspace(0, length_seconds, length_seconds * sample_freq + 1)
signal = np.sin(2 * np.pi * time * signal_frequency) + np.random.random(size=time.size) * noise_amplitude
```

```python
def calc_fft(sample_rate, signal):
    # Compute the FFT freq. bins
    frq = rfftfreq(signal.size, d=1./sample_rate)

    # Compute FFT and normalize is two ways
    X = rfft(signal)
    X_len_norm = X/(0.5*signal.size)
    return frq, X_len_norm
```

```python
fft_x, fft_y = calc_fft(sample_freq, signal)
```

```python
fig = plt.figure(figsize=(12,9))
ax1 = plt.subplot(111)

# Set labels and tick sizes
ax1.set_xlabel(r'Time [sec]',fontsize=18)
ax1.set_ylabel(r'Voltage',fontsize=18)
ax1.tick_params(axis='both', which='major', labelsize=16)
ax1.grid()

# Plotting
ax1.plot(time, signal)

#plt.legend(frameon=False, fontsize=16)
ax1.set_xlim(0, 1)
#ax1.set_ylim(1e-6, 2)
plt.savefig('scope_zoomout.png', bbox_inches="tight"); 
```

```python
fig = plt.figure(figsize=(12,9))
ax1 = plt.subplot(111)

# Set labels and tick sizes
ax1.set_xlabel(r'Frequency',fontsize=18)
ax1.set_ylabel(r'|Y(freq)|',fontsize=18)
ax1.tick_params(axis='both', which='major', labelsize=16)
ax1.grid()

# Plotting
ax1.plot(fft_x, np.abs(fft_y), color='tab:purple')

#ax1.set_xlim(0.001, 200)
#ax1.set_ylim(1e-6, 2)
plt.savefig('spectrum_linear.png', bbox_inches="tight"); 
```
